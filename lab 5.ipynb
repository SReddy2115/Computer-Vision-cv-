```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image and convert it to grayscale
image = cv2.imread('s2.jpg')
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

#https://learnopencv.com/otsu-thresholding-with-opencv/  otsus thresholding
#https://stackoverflow.com/questions/70574748/what-settings-in-cv2-threshold-to-threshold-this-image
_, binary = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

#points for counting 
contours, _ = cv2.findContours(binary, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

dotcount = 0
white_color = (255, 255, 255)
red_color = (0, 0, 255)

#counting dots
for contour in contours:
    if cv2.contourArea(contour) > 80:  # this dection is working only beteen 80-180
        dotcount += 1
        x, y, w, h = cv2.boundingRect(contour)
        center_x, center_y = x + w // 2, y + h // 2
        cv2.rectangle(image, (x, y), (x + w, y + h), white_color, 2)
        cv2.circle(image, (center_x, center_y), 5, red_color, -1)

plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()

print(f"Total OLED Centers Detected: {dotcount}")
```
