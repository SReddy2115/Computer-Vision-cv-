# 2. Experiment with the drawing functions in the imshow window. Create some interesting design / game interface using the openCV functions.

# 3. Experiment with keyboard events. How to listen and identify key events and create animation using loops in the imshow window. 

```python
import cv2
import numpy as np
import random

# Constants
WINDOW_SIZE = 1000
GRID_SIZE = 20
SNAKE_SIZE = 20

# Initialize the game
snake = [(100, 100)]
food = (200, 200)
direction = (1, 0)
score = 0

# Create a blank canvas
canvas = np.zeros((WINDOW_SIZE, WINDOW_SIZE), dtype=np.uint8)

while True:
    # Clear the canvas
    canvas.fill(0)

    # Draw the snake
    for segment in snake:
        x, y = segment
        canvas[y:y + SNAKE_SIZE, x:x + SNAKE_SIZE] = 255

    # Draw the food
    fx, fy = food
    canvas[fy:fy + SNAKE_SIZE, fx:fx + SNAKE_SIZE] = 128

    # Display the score
    cv2.putText(canvas, f"Score: {score}", (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, 255, 2)

    # Display the canvas
    cv2.imshow('Snake Game', canvas)

    # Capture keyboard input
    key = cv2.waitKey(100)

    # Update the direction based on keyboard input
    if key == 38:
        direction = (0, -1)
    elif key == 40:
        direction = (0, 1)
    elif key == 37:
        direction = (-1, 0)
    elif key == 39:
        direction = (1, 0)
    elif key == 27:  # Exit when Esc key is pressed
        break

    # Update the snake's position
    head = snake[-1]
    new_head = (head[0] + direction[0] * SNAKE_SIZE, head[1] + direction[1] * SNAKE_SIZE)
    snake.append(new_head)

    # Check if the snake eats the food
    if new_head == food:
        score += 1
        food = (random.randrange(0, WINDOW_SIZE - SNAKE_SIZE, GRID_SIZE), random.randrange(0, WINDOW_SIZE - SNAKE_SIZE, GRID_SIZE))
    else:
        snake.pop(0)

    # Check for game over conditions
    if (new_head[0] < 0 or new_head[0] >= WINDOW_SIZE or
            new_head[1] < 0 or new_head[1] >= WINDOW_SIZE or
            new_head in snake[:-1]):
        print("Game Over")
        break

# Close the window
cv2.destroyAllWindows()
```
